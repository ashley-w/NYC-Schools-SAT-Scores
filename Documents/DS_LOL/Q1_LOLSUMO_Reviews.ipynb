### Final Attempt:

I added a language column using the values from the detect_language function.

```python
lang = []

for row in q1_reviews_cl['content']:
    if len(row) > 3:
        var = TextBlob(row)
        language = var.detect_language()
        lang.append(language)
    else:
        lang.append('Unknown')
    
# print(lang)
q1_reviews_cl['lang'] = lang
q1_reviews_cl
```

### Creating a List with Translated Reviews

Much like the language list, I'm starting with a list so I can check out the contents. 

```python
tr_review_txt = []

for row in q1_reviews_cl['content']:
    if len(row) > 4:
        var = TextBlob(row)
        if var.detect_language() != 'en':
            trans = var.translate(to='en')
            tr_review_txt.append(trans)
        else: # This is
            tr_review_txt.append(row)
        
print(tr_review_txt)
```
